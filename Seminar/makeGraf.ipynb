# Creating a graf 

```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file
# Skip the first 3 rows and the last 2 rows, and set the delimiter to a series of spaces
df = pd.read_csv('./meritve/city.txt', skiprows=2, skipfooter=3, delim_whitespace=True, engine='python')

print(df)
```

```python
# Plot 'Hist_GPU' as a bar plot in blue
plt.bar(df['Beam'], df['Hist_GPU'], color='blue', width=1, label='Hist_GPU')

# Plot 'New_Hist' as a bar plot in red, shift the 'Beam' values by 0.4 to avoid overlap
plt.bar(df['Beam'], df['New_Hist'], color='red', width=1, label='New_Hist')

plt.xlabel('Beam (Color)')
plt.ylabel('Value')
plt.title('Color Frequency in Hist_GPU and New_Hist')
plt.legend()

# Disable scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.xlim(0, 255)

plt.show()
```

```python
# Normalize 'CDF_GPU' and 'New_CDF' data
normalized_cdf_gpu = df['CDF_GPU'] / df['CDF_GPU'].max()
normalized_new_cdf = df['New_CDF'] / df['New_CDF'].max()

# Plot normalized 'CDF_GPU' data
plt.plot(normalized_cdf_gpu, linestyle='-', color='blue', label='CDF_GPU')

# Plot normalized 'New_CDF' data
plt.plot(normalized_new_cdf, linestyle='-', color='red', label='New_CDF')

plt.xlabel('Index')
plt.ylabel('Normalized Value')
plt.title('Normalized CDF_GPU and New_CDF Data')
plt.legend()
plt.grid(True)

# Set the range of y-axis from 0 to 1
plt.ylim(0, 1.1)

# Disable scientific notation
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
